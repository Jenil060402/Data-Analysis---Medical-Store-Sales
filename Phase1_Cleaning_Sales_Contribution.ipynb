# Phase 1: Data Cleaning, Pre-Processing

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('Sales_Contribution.xlsx', names = ['Sr', 'Comp', 'Code', 'Item Description','', 'Packing', 'Stock', 'Qnty', 'Sales', 'Contri%'], skiprows=4)
ndf = df
ndf
```

```python
ndf.columns
```

```python
ndf.drop(['', 'Sr'], axis=1, inplace=True)
ndf
```

```python
ndf.dropna(subset=['Item Description'], inplace=True)
ndf
```

*The Contri% column seems inconsistent with the available data, hence dropping it entirely seems a good idea*

```python
ndf.drop('Contri%', axis = 1, inplace=True)
ndf
```

*Now, I will find the items with 0 potential sales, or negligible Qnty, and then decide what to do.*

```python
ndf[ndf.Sales.isna()]
```

*Filling these with 0 seems a better option, but first, I will find whether there are other missing values*

```python
ndf[ndf['Item Description'].isna()]
```

```python
ndf.isna().sum()
```

```python
ndf[ndf['Qnty'].isna()]
```

*I got to know that stock column is missing 335 values, and filling all with 0 is a dumb decision, a better decision is to fill them with same value as Qnty.*

```python
ndf['Stock'].fillna(ndf['Qnty'], inplace=True)
ndf
```

```python
ndf.isna().sum()
```

```python
ndf[ndf.Qnty.isna()]
```

```python
ndf[ndf.Stock.isna()]
```

*The null values in Qnty is a lot, and I will use regression to later deal with them, for now, let's remove other null values*

```python
ndf.dropna(subset=['Stock', 'Sales'], inplace = True, how = 'all')
```

```python
ndf[ndf.Stock.isna()]
```

```python
ndf.Stock.fillna(1,inplace=True)
```

```python
ndf[ndf.Stock.isna()]
```

```python
ndf.fillna({'Packing': '1 Pack'}, inplace = True)
```

```python
ndf.isna().sum()
```

```python
ndf.Sales.fillna(0, inplace = True)
```

```python
ndf.isna().sum()
```

```python
ndf.to_excel('clean_salescontri.xlsx', index = False)
```
