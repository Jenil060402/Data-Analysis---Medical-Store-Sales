*The data was segregated into categories by parititioning the data row wise, the below function enters the category into 'category' column through indexing.*

```python
def assign_category(index):
    if 0 <= index < 2299:
        return "Allopathic"
    elif 2301 <= index < 2436:
        return "Ayurvedic"
    elif 2437 <= index < 2603:
        return "Cosmetics"
    elif 2604 <= index < 2641:
        return "Dental"
    elif 2642 <= index < 2651:
        return "Condom"
    elif 2652 <= index < 2834:
        return "Food Item"
    elif 2835 <= index < 3093:
        return "General OTC Ext."
    elif 3094 <= index < 3179:
        return "General OTC Int."
    elif 3180 <= index < 3209:
        return "Child Product"
    elif 3210 <= index < 3259:
        return "Generic"
    elif 3260 <= index < 3262:
        return "Homeopathic"
    elif 3263 <= index < 3265:
        return "I.V. Fluid"
    elif 3266 <= index < 3273:
        return "Insecticide"
    elif 3274 <= index < 3287:
        return "Women's Sanitary"
    elif 3288 <= index < 3368:
        return "Surgical Item"
    elif 3369 <= index < 3404:
        return "Toiletry"
    elif 3405 <= index < 3408:
        return "Veterinary"
    else:
        return None

```

```python
ndf['Category'] = ndf.index.map(assign_category)
```
