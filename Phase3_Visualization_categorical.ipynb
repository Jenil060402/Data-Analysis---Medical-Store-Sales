# Phase 3: Visualization and Analytics - Categorical Sales

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_excel('clean_categorical_sales.xlsx')
df1
```

**Finding total Sales and total Qnty sold for each category**

```python
cat_sales = df1.groupby('Category').sum().reset_index()
cat_sales
```

```python
plt.figure(figsize=(10, 10), dpi=350)
sns.barplot(data = cat_sales, x = 'Category', y = 'Sales')
plt.xticks(rotation = 45)
plt.title('Bar Chart for Sales across Categories(including Allopathic)')
plt.savefig('Bar Chart for Sales across Categories(including Allopathic).jpg', format= 'jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 10), dpi=350)
sns.barplot(data = cat_sales[1:], x = 'Category', y = 'Sales')
plt.xticks(rotation = 45)
plt.title('Bar Chart for Sales across Categories(excluding Allopathic)')
plt.savefig('Bar Chart for Sales across Categories(excluding Allopathic).jpg', format= 'jpg')
plt.show()
```

```python
cat_avg_sales = df1.groupby('Category')[['Qnty', 'Sales']].mean()
cat_avg_sales
```

```python
plt.figure(figsize=(12, 7), dpi=350)
sns.barplot(data = cat_avg_sales, x = 'Category', y = 'Sales', hue = 'Qnty', palette = 'flare_r')
sns.cubehelix_palette(rot=.9)
plt.xticks(rotation = 60)
plt.ylabel('Sales in â‚¹')
plt.title('Avg. Sale & Qnty per Category.jpg')
plt.savefig('Avg. Sale & Qnty per Category.jpg', format = 'jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 10), dpi=350)
sns.barplot(data = cat_sales, x = 'Category', y = 'Qnty')
plt.xticks(rotation = 45)
plt.title('Bar Chart for Quantity across Categories(including Allopathic)')
plt.savefig('Bar Chart for Quantity across Categories(including Allopathic).jpg', format= 'jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 10), dpi=350)
sns.barplot(data = cat_sales[1:], x = 'Category', y = 'Qnty')
plt.xticks(rotation = 45)
plt.title('Bar Chart for Quantity across Categories(excluding Allopathic)')
plt.savefig('Bar Chart for Quantity across Categories(excluding Allopathic).jpg', format= 'jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 10), dpi=350)
plt.pie(cat_sales.Sales, labels = cat_sales.Category.unique(), rotatelabels = 45, autopct = '%.1f%%', radius=.8)
plt.title('Pie Chart for Sales % across Categories(including Allopathic)')
plt.savefig('Pie Chart for Sales % across Categories(including Allopathic).jpg', format= 'jpg')
plt.show()
```

```python
plt.figure(figsize=(10, 10), dpi=350)
plt.pie(cat_sales.Sales[1:], labels = cat_sales.Category.unique()[1:], rotatelabels = 45, autopct = '%.1f%%', radius =.8)
plt.title('Pie Chart for Sales % across Categories(excluding Allopathic)')
plt.savefig('Pie Chart for Sales % across Categories(excluding Allopathic).jpg', format= 'jpg')
plt.show()
```
